```python
# Step 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load the data
url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
df = pd.read_csv(url, parse_dates=['date'])

# Step 3: Filter and clean
df = df[['location', 'date', 'new_cases', 'new_deaths', 'total_vaccinations', 'people_fully_vaccinated']]
df.dropna(subset=['location'], inplace=True)

# Step 4: Choose a country to visualize
country = "New Zealand"  # change to your desired country
country_df = df[df['location'] == country].sort_values('date')

# Step 5: Plot daily cases and deaths
plt.figure(figsize=(12, 6))
sns.lineplot(data=country_df, x='date', y='new_cases', label='New Cases')
sns.lineplot(data=country_df, x='date', y='new_deaths', label='New Deaths')
plt.title(f'COVID-19: New Cases & Deaths in {country}', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=10, fontweight='bold')
plt.ylabel('Count', fontsize=10, fontweight='bold')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 6: Plot vaccinations
plt.figure(figsize=(12, 6))
sns.lineplot(data=country_df, x='date', y='total_vaccinations', label='Total Vaccinations')
sns.lineplot(data=country_df, x='date', y='people_fully_vaccinated', label='Fully Vaccinated')
plt.title(f'COVID-19: Vaccination Progress in {country}', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=10, fontweight='bold')
plt.ylabel('People', fontsize=10, fontweight='bold')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
